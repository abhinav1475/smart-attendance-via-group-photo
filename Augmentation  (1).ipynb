```python
import os
import cv2
import numpy as np
import albumentations as A
from PIL import Image

# Paths
input_path = '/content/drive/MyDrive/Mini/faces'  # Folder containing cropped faces
output_path = '/content/drive/MyDrive/Mini/augmented_faces'  # Folder to save augmented images

# Ensure output directory exists
os.makedirs(output_path, exist_ok=True)

# Augmentation pipeline with multiple augmentations
transform = A.Compose([
    A.HorizontalFlip(p=0.5),
    A.RandomRotate90(p=0.5),
    A.RandomBrightnessContrast(p=0.5),
    A.RandomScale(p=0.5),
    A.GaussianBlur(p=0.3),
    A.ShiftScaleRotate(p=0.5),
])

# Iterate through each student's folder and apply augmentation
for student_folder in os.listdir(input_path):
    student_folder_path = os.path.join(input_path, student_folder)
    if not os.path.isdir(student_folder_path):
        continue

    # Create a folder for augmented images
    student_output_folder = os.path.join(output_path, student_folder)
    os.makedirs(student_output_folder, exist_ok=True)

    # Iterate through each face image in the student's folder
    for img_name in os.listdir(student_folder_path):
        if not img_name.lower().endswith('.jpg'):
            continue

        img_path = os.path.join(student_folder_path, img_name)

        # Load the image
        img = cv2.imread(img_path)

        if img is None:
            print(f"[ERROR] Cannot read image: {img_name}")
            continue

        # Save the original image first
        original_img_name = f"{os.path.splitext(img_name)[0]}_original.jpg"
        original_img_path = os.path.join(student_output_folder, original_img_name)
        cv2.imwrite(original_img_path, img)
        print(f"[OK] Saved original {original_img_path}")

        # Generate 10 augmented images with random transformations
        for i in range(10):
            augmented = transform(image=img)
            augmented_img = augmented['image']

            # Save augmented images with unique names
            augmented_img_name = f"{os.path.splitext(img_name)[0]}_aug_{i+1}.jpg"
            augmented_img_path = os.path.join(student_output_folder, augmented_img_name)

            # Save augmented image
            cv2.imwrite(augmented_img_path, augmented_img)
            print(f"[OK] Saved {augmented_img_path}")

print("Augmentation complete.")
```
