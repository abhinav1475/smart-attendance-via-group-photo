```python
import os
from PIL import Image
import cv2
import numpy as np
from mtcnn import MTCNN

# Input: high-quality roll.jpg images
input_path = '/content/drive/MyDrive/Mini/imgs'
output_path = '/content/drive/MyDrive/Mini/faces'

# Ensure output directory exists
os.makedirs(output_path, exist_ok=True)

# Initialize MTCNN face detector
detector = MTCNN()

for img_name in os.listdir(input_path):
    if not img_name.lower().endswith('.jpg'):
        continue

    try:
        # Full path to input image
        img_path = os.path.join(input_path, img_name)

        # Load image and convert to RGB
        img = cv2.imread(img_path)
        if img is None:
            print(f"[ERROR] Cannot read image: {img_name}")
            continue

        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # Detect faces using MTCNN
        results = detector.detect_faces(img_rgb)
        if not results:
            print(f"[WARNING] No face found in {img_name}")
            continue

        # Use the most confident face
        best_face = max(results, key=lambda x: x['confidence'])
        x, y, w, h = best_face['box']

        # Apply margin
        margin = 0.2
        dx = int(w * margin)
        dy = int(h * margin)
        x1 = max(x - dx, 0)
        y1 = max(y - dy, 0)
        x2 = min(x + w + dx, img_rgb.shape[1])
        y2 = min(y + h + dy, img_rgb.shape[0])

        # Crop face
        face_crop = img_rgb[y1:y2, x1:x2]

        # Check if the face crop is valid (non-empty)
        if face_crop.size == 0:
            print(f"[ERROR] Empty face crop for {img_name}")
            continue

        # Convert to PIL for saving
        face_pil = Image.fromarray(face_crop)

        roll_id = os.path.splitext(img_name)[0]
        roll_dir = os.path.join(output_path, roll_id)
        os.makedirs(roll_dir, exist_ok=True)

        # Ensure face is saved with unique filename
        out_path = os.path.join(roll_dir, 'face_1.jpg')

        # Save image with high quality (do not resize here)
        face_pil.save(out_path, format='JPEG', quality=100)

        print(f"[OK] Saved {out_path}")

    except Exception as e:
        print(f"[ERROR] Failed {img_name}: {e}")
```
